```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

# Assuming you have loaded your data into the DataFrame 'df'
# For example, if 'Species' is the target variable, and the rest are features
features = df.drop('Species', axis=1)  # Assuming 'Species' is the target variable
target = df['Species']

# Split the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(features, target, test_size=0.2, shuffle=True)

# Check the shape of the resulting sets
print("Shape of x_train:", x_train.shape)
print("Shape of x_test:", x_test.shape)
print("Shape of y_train:", y_train.shape)
print("Shape of y_test:", y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
lm=LogisticRegression()
lm.fit(x_train,y_train)
```

```python
y_pred = lm.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
y_pred = lm.predict(x_test)
accuracy_score(y_test,y_pred)
```

```python
models = ['Logistic Regression']
accuracy = [1.0 ] 

plt.figure(figsize=(10, 6))
plt.bar(models, accuracy, color=['skyblue'])
plt.xlabel('Model', color='red')
plt.ylabel('Accuracy',color='red')
plt.title('Accuracy of Model is 100 % ',color='green')
plt.ylim(0.70, 1.50)
for i, acc in enumerate(accuracy):
    plt.text(i, acc + 0.001, f'{acc:.4f}', ha = 'center')

plt.show()
```
